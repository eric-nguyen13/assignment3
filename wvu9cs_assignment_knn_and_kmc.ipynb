````python
# Create a matrix X of Age and AnnualSalary and an outcome y equaling Purchased
vars = ['Age','AnnualSalary']
X = df.loc[:,vars]
y = df['Purchased']
````

````python
# 3. MaxMin-normalize Age and AnnualSalary in X
def maxmin(x): # Maxmin normalization function
  x = (x-min(x))/(max(x)-min(x))
  return x

X = X.apply(maxmin) # Apply maxmin normalization function to all columns of X
X.describe()
````

````python
# 4. Split the sample into a ~80% training and ~20% testing dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2,random_state=42)
````

````python
# 5. Determine the optimal number of neighbors k to use
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier # Import the kNN classifier tool

# Determine the optimal k:
k_bar = 30
k_grid = np.arange(1,k_bar) # The range of k's to consider
SSE = np.zeros(k_bar) # Object to save the Sum of Squared Error
for k in range(k_bar):
  model = KNeighborsClassifier(n_neighbors=k+1) # Create a sk model for k
  fitted_model = model.fit(X_train,y_train) # Train the model on our data
  y_hat = fitted_model.predict(X_test) # Predict values for test set
  SSE[k] = np.sum( (y_test-y_hat)**2 ) # Save the computed SSE

SSE_min = np.min(SSE) # Lowest recorded SSE
min_index = np.where(SSE==SSE_min) # Find the indices of y that equal the minimum
k_star = k_grid[min_index] # Find the optimal value of k
print("star: " + str(k_star))

plt.plot(np.arange(0,k_bar),SSE,label='Test') # Plot SSE by k
plt.xlabel("k")
plt.ylabel("SSE")
plt.legend(loc='lower right')
plt.title('SSE')
plt.show()
````

````python
# 6. Cross tabulate (Confusion Matrix)
model = KNeighborsClassifier(n_neighbors=9) # Create a sk model for k
fitted_model = model.fit(X_train,y_train) # Train the model on our data
y_hat = fitted_model.predict(X_test) # Predict values for test set
pd.crosstab(y_test, y_hat)
````

````python
# 1. Load the ./data/USA_cars_datasets.csv . Keep selected variables. Are there any NAs? Look at head and dimensions.
df2 = pd.read_csv('./assignment3/data/USA_cars_datasets.csv')
print("\nDimensions: " + str(df2.columns) + "\n")
df2.head()

# Keep only the specified variables (price, year, mileage) and drop the rest
selected_columns = ['price', 'year', 'mileage']
df2 = df2[selected_columns]

# Check for missing values
missing_values = df2.isna().sum()

print("\nMissing values: \n" + str(missing_values))
print("\nDimensions: " + str(df2.columns) + "\n")
print("Shape: " + str(df2.shape) + "\n")
df2.head()
````

There are no NA's to handle here.

````python
# 2. Maxmin normalize year and mileage.
def maxmin(x): # Maxmin normalization function
    x = (x-min(x))/(max(x)-min(x))
    return x

vars = ['year','mileage']
X = df2.loc[:,vars]
y = df2['price']


X = X.apply(maxmin) # Apply maxmin normalization function to all columns of X
X.describe()
````

````python
# 3. Split the sample into ~80% for training and ~20% for evaluation.

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2,random_state=42)
````

````python
# 4. Use the   NN algorithm for regression and the training data to predict price using year and mileage for the test set for  =3,10,25,50,100,300 .
# For each value of   , compute the Sum of Squared Error and make a scatterplot showing the test value plotted against the predicted value.
# What patterns do you notice as you increase   ?

from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

k_values = [3, 10, 25, 50, 100, 300]

sse_values = {}

for k in k_values:
    knn_regressor = KNeighborsRegressor(n_neighbors=k)

    knn_regressor.fit(X_train, y_train)

    y_pred = knn_regressor.predict(X_test)

    # Calculate the SSE
    sse = np.sum((y_test - y_pred) ** 2)
    sse_values[k] = sse

    # Create a scatterplot for the predicted vs. actual 'price' for the current k
    plt.figure(figsize=(8, 6))
    plt.scatter(y_test, y_pred, alpha=0.5)
    plt.title(f'Scatterplot for k = {k}')
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.show()

# Print SSE
for k, sse in sse_values.items():
    print(f'k = {k}: SSE = {sse}')
````

````python
# 5. Determine the optimal    for these data.

k_values = [3, 10, 25, 50, 100, 300]

SSE_values = []  # Create a list to store SSE values

for k in k_values:
    model = KNeighborsRegressor(n_neighbors=k)
    fitted_model = model.fit(X_train, y_train)
    y_hat = fitted_model.predict(X_test)
    SSE = np.sum((y_test - y_hat) ** 2)
    SSE_values.append(SSE)

optimal_k = k_values[np.argmin(SSE_values)]  # Find the k with the lowest SSE

print(f"Optimal k: {optimal_k}")
print(SSE_values)

plt.plot(k_values, SSE_values, marker='o')
plt.xlabel("k")
plt.ylabel("SSE")
plt.title('SSE for Different k Values')
plt.grid(True)
plt.show()

````

6. When analyzing the Actual vs. Predict Prices charts, there are patterns that can be observed. Specifically, when there are low values of k (such as 3 and 10), the model is highly influenced by data points so there is overfitting. Contrastingly with high values of k (such as 300), there is underfitting because the model is too simple and doesn't detect the true patterns. When K is optimal (in this case, k = 50), there is no overfitting or underfitting. Based on the chart above, it is clear that k=50 is optimal as it has the lowest SSE.

2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?

````python
# 1. Run the code below

# Code Given
import numpy as np
import pandas as pd

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

Based on my observations, it's clear that as the noise increases from 0.125 to 2.0, the distinction between the clusters disappears. Noise contributes to increased uncertainty within KMeans, causing the visuals to begin blending the clusters together. Noise = 0.5 is the last level in which the clusters can be distinguished apart. Noise levels after that make it unintelligible.

````python
# 3. Create a scree plot for each of the datasets.
# Describe how the level of noise affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a   .

from sklearn.cluster import KMeans

def maxmin(x): # Max-min scaler function
    x = (x-min(x))/(max(x)-min(x))
    return x

def scree(data): # I built a function to make scree plots quickly
    # Preprocess the data:
    X = data.loc[ : , ['x1','x2'] ] # Select data to cluster
    X = X.apply(maxmin) # Normalize the data
    # Create a scree plot:
    k_bar = 15
    k_grid = np.arange(1,k_bar+1) # The range of k's to consider
    SSE = np.zeros(k_bar) # We'll store the training error here
    for k in range(k_bar): # For indices from 0 to k_bar-1
        model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0) #
        model = model.fit(X) # Fit the emodel
        SSE[k] = model.inertia_ # Record the sum of squared error for k in the SSE
    scree_plot, axes = plt.subplots()
    sns.lineplot(x=k_grid, y=SSE).set_title('Scree Plot') # Create scree plot
    axes.set_ylim(0, 35)
````

The scree plot shows a large drop indicating the optimal value for k. However, when noise levels increase, the distinction between clusters begins to disappear and there is not as large of a drop. The intuition behind the elbow is that when the data is distinct, the scree plot will indicate a large drop in sse. However, when the data is noisy/mixed, it's difficult to discern how many clusters to use and which cluster to assign the points to. This means that the scree plot will show a smooth curve and not indicate a good k value. It is an excellent tool to have when your data is organized.
